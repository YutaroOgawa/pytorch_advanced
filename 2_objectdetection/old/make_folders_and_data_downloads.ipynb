# 「第2章 物体検出」の準備ファイル

- 本ファイルでは、第2章で使用するフォルダの作成とファイルのダウンロードを行います。

```python
import os
import urllib.request
import zipfile
import tarfile
```

```python
# フォルダ「data」が存在しない場合は作成する
data_dir = "./data/"
if not os.path.exists(data_dir):
    os.mkdir(data_dir)
```

```python
# フォルダ「weights」が存在しない場合は作成する
weights_dir = "./weights/"
if not os.path.exists(weights_dir):
    os.mkdir(weights_dir)
```

```python
# VOC2012のデータセットをここからダウンロードします
# 時間がかかります
url = "http://host.robots.ox.ac.uk/pascal/VOC/voc2012/VOCtrainval_11-May-2012.tar"
target_path = os.path.join(data_dir, "VOCtrainval_11-May-2012.tar") 

if not os.path.exists(target_path):
    urllib.request.urlretrieve(url, target_path)
    
    tar = tarfile.TarFile(target_path)  # tarファイルを読み込み
    tar.extractall(data_dir)  # tarを解凍
    tar.close()  # tarファイルをクローズ
    
```

```python
# 学習済みのSSD用のVGGのパラメータを./weights/にダウンロード
# MIT License
# Copyright (c) 2017 Max deGroot, Ellis Brown
# https://github.com/amdegroot/ssd.pytorch
    
url = "https://s3.amazonaws.com/amdegroot-models/vgg16_reducedfc.pth"
target_path = os.path.join(weights_dir, "vgg16_reducedfc.pth") 

if not os.path.exists(target_path):
    urllib.request.urlretrieve(url, target_path)
```

```python
import os
import urllib.request
import tarfile
#import zipfile

#------------------
# フォルダやファイルがないときに、作成したりダウンロードしたりする関数を作成
#------------------
def make_target_dir(target_dir):
    if not os.path.exists(target_dir):
        os.mkdir(target_dir)

def download_target_file(url, target_path):
    if not os.path.exists(target_path):
        urllib.request.urlretrieve(url, target_path)

def download_target_file_and_untar(url, target_path, extra_path):
    if not os.path.exists(target_path):
        urllib.request.urlretrieve(url, target_path)
        # tarファイルを読み込み
        tar = tarfile.TarFile(target_path)

        tar.extractall(extra_path)  # tarを解凍
        tar.close()  # tarファイルをクローズ
```

```python
# 学習済みSSDの重みを./weights/に置いておく
url = "https://s3.amazonaws.com/amdegroot-models/ssd300_mAP_77.43_v2.pth"
target_path = os.path.join(weights_dir, "ssd300_mAP_77.43_v2.pth") 
download_target_file(url, target_path)
```

```python
# 学習済みvggの重みを./weights/に置いておく
url = "https://s3.amazonaws.com/amdegroot-models/vgg16_reducedfc.pth"
target_path = os.path.join(weights_dir, "vgg16_reducedfc.pth") 
download_target_file(url, target_path)
```
